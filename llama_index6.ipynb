Joint QA Summary Engine

```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader("data").load_data()
```

```python
documents
```

```python
from llama_index.llms.openai import OpenAI

gpt = OpenAI(temperature=0.8, model= "gpt-4o-mini")

chatgpt = OpenAI(temperature=0.8,model="gpt-3.5-turbo")
```

```python
from llama_index.core.composability import QASummaryQueryEngineBuilder

# can also specify an existing docstore, summary text, qa_text, etc.
query_engine_builder = QASummaryQueryEngineBuilder(
    llm=gpt,
)
query_engine = query_engine_builder.build_from_documents(documents)
```

```python
response = query_engine.query("can you give me a summary of the author's life?")
```

```python
print(response)
```

```python
res = query_engine.query("can you give me some story for life?")
```
